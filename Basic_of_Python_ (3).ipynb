2. What is an interpreter in Python?

3. What are pre-defined keywords in Python?

   - Pre-defined keywords are reserved words in Python used to define syntax and structure (e.g., `if`, `else`, `while`, `def`). They cannot be used as variable names.



 4. Can keywords be used as variable names?

 5. What is mutability in Python?


 - Mutability means whether the value of an object can be changed after creation.

  -  Mutable: Can change (e.g., `list`, `dict`)
  -  Immutable: Cannot change (e.g., `int`, `str`, `tuple`)


 6. Why are lists mutable but tuples are immutable?


   - Lists allow item changes (e.g., adding/removing), while tuples are fixed. This helps with performance and ensures data safety.



 7. What is the difference between "==" and " is
 " operators in Python?


  * `==` checks **value equality** (if values are the same)
   * `is` checks **identity** (if both refer to the same object in memory)


 8. What are logical operators in Python?

- Logical operators combine conditional statements:

* `and` – True if both are true
* `or` – True if at least one is true
* `not` – Reverses the result



9. What is type casting in Python?


   - Type casting means converting one data type into another (e.g., `int("10")`, `str(5)`).



 10. What is the difference between implicit and explicit type casting?
*

* **Implicit:** Python automatically converts (e.g., `int + float → float`)
* **Explicit:** Programmer manually converts (e.g., `int("5")`)



 11. What is the purpose of conditional statements in Python?


They help run specific code when a condition is met using `if`, `elif`, and `else`.



 12. How does the else statement work?


`else` executes a block of code if all previous `if`/`elif` conditions are false.



13. What is the difference between `for` and `while` loops?


* `for`: Repeats a block for a defined number of times.
* `while`: Repeats as long as the condition is true (may be infinite).
